```python
import os
import pandas as pd

# 获取CSV文件地址
path = os.path.abspath('./../../../Traffic_Simulation_Data_Generation_for_Baselines/0/') 
data_path = os.path.join(path, 'TraCI_output_adjusted.csv')

# 载入CSV文件
df = pd.read_csv(data_path)
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# 绘制 'Wait_Time' 的数据分布图
plt.figure(figsize=(8, 6))
plt.hist(df['Wait_Time'], bins=100, edgecolor='black')
plt.title('Distribution of Wait_Time')
plt.xlabel('Wait_Time')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 计算 'Wait_Time' 列的统计数据
wait_time_stats = {
    'Max': df['Wait_Time'].max(),
    'Min': df['Wait_Time'].min(),
    'Mean': df['Wait_Time'].mean(),
    'Median': df['Wait_Time'].median(),
    'Variance': df['Wait_Time'].var(),
    'Standard Deviation': df['Wait_Time'].std()
}

print(wait_time_stats)
```
